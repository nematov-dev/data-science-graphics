```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Dataset

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

scatterplot()

```python
sns.scatterplot(data=df, x="year", y="price")
plt.show()
```

1970 yildagi ma'lumotni tashlab yuboramiz

```python
df = df[df.year != 1970 ]
sns.scatterplot(data=df, x="year", y="price")
plt.show()
```

Karralatsiya ko'rish uchun chiziq chizamiz



```python
sns.regplot(data=df, x="year", y="price")
plt.show()
```

Qo'shimcha metodlar

```python
sns.scatterplot(data=df, x="year", y="price", hue="transmission",style='fuelType')  # hue ma'lumotlarni ranglar bilan ajratish, style ma'lumotlarni shakllar bilan ajratish
plt.show()
```

Swarmplot

```python
# swarmplot o'rniga stripplot
sns.stripplot(data=df, x='fuelType', y='price', jitter=True, alpha=0.5)
plt.show()
```
