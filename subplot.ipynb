```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

sin(x) , cos(x) grafiklarini subplot bilan birlashtirib chizamiz

```python
x = np.arange(0,4*np.pi,0.1)

sin = np.sin(x)
cos = np.cos(x)
```

Grafik hajmini ko'rsatamiz

```python
plt.figure(figsize=(12,6))
```

Subplot


```python
# 1 - grafik uchun subplot
plt.subplot(2,1,1)

plt.plot(
    x,sin,
    linewidth=1.5,
    linestyle="-.",
    color="red",
    alpha=0.7,
    marker="h",
)
plt.title('six (x) funksiya grafigi')
plt.grid(color = 'tomato',alpha=0.5,linestyle='--')


# 2 - grafik uchun subplot

plt.subplot(2,1,2)

plt.plot(
    x,cos,
    linewidth=1.5,
    linestyle=":",
    color="c",
    alpha=0.7
)
plt.title('cos (x) funksiya grafigi')
plt.grid(color = 'tomato',alpha=0.5,linestyle='--')

plt.show()
```

```python
sns.barplot(data=df,x='year',y='price')
plt.xticks(rotation=90)
plt.show()
```

Subplots

```python
fig,ax = plt.subplots(1,2,figsize=(15,5)) # subplots

sns.scatterplot(ax=ax[0],data=df,x='year',y='mileage',hue='price',style='fuelType') # 1-grafik
plt.grid()

sns.barplot(ax=ax[1],data=df,x='year',y='price') # 2-grafik

plt.xticks(rotation=90)

fig.suptitle('Scatterplot va Barplot')

plt.show()
```
