```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sonlar = np.arange(20)
plt.plot(sonlar)
plt.show()
```

```python
from matplotlib.lines import lineStyles
x = np.arange(10)
y = x**2
plt.plot(x, y,color='r',linestyle='--',linewidth=5)
plt.grid()
plt.title("Grafik")
plt.show()
```

```python
x= np.arange(0,4*np.pi,0.1)
sin = np.sin(x)
cos = np.cos(x)
```

```python
plt.figure(figsize=(12,6))
```

```python
plt.plot(x,sin,linewidth=1.5,linestyle='-',color='r',alpha=0.7,marker='h')
plt.plot(x,cos,linewidth=1.5,linestyle=':',color='c',alpha=0.7,marker='v')
```

```python
plt.plot(x,sin,linewidth=1.5,linestyle='-',color='r',alpha=0.7,marker='h')
plt.plot(x,cos,linewidth=1.5,linestyle=':',color='c',alpha=0.7,marker='v')
plt.title('sin(x) va cos(x)')
plt.xlabel('x')
plt.legend(['sin(x)','cos(x)'],loc=0,fontsize='x-large')
plt.grid(color='tomato',alpha=0.5)
plt.savefig('grafik.png')
plt.show()
```
